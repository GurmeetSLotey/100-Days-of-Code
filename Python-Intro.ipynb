## Scope & Number guessing game

### Namespace Local and Global Scope

Variables inside a function are local scope

Variables outside of functions are global scope

```python
enemies = 1

def increase_enemies():
    enemies = 2
    print(f"enemies inside function: {enemies}")

increase_enemies()
print(f"enemies outside function: {enemies}")
```

```python
global enemies # while enemies does not need global as it is already a global variable
enemies = 1


# variable inside the function are local
def increase_enemies():
    enemies = 2
    print(f"enemies inside function: {enemies}")

increase_enemies()
print(f"enemies outside function: {enemies}")
```

```python
enemies = 1

def increase_enemies():
    global enemies # this is how we modify the global scope enemies
    enemies = 2
    print(f"enemies inside function: {enemies}")

increase_enemies()
print(f"enemies outside function: {enemies}")
```

```python
enemey = 1

def increase_enemies():
    print(f"enemies inside function: {enemey}")

increase_enemies()
print(f"enemies outside function: {enemey}")
```

```python
class User:

    def __init__(self, user_id, username):
        self.id = user_id
        self.username = username
        self.followers = 0 # default value
        self.following = 0

    def follow(self, user):
        user.followers += 1 # we pass in the we are now following another user thus their follower count increases
        self.following += 1 # we use self as we started following someone
        print(f"{self.username} has started following {user.username}")

user_1 = User("007","JamesBond")
user_2 = User("000", "Dr. Evil")

print(user_1.id, user_1.username)
print(f"User_2 followers: {user_2.followers}")

user_1.follow(user_2)

print(f"User_2 followers: {user_2.followers}")
```

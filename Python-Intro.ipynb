### Modulo Operator %

It works out what is the remainder after the divison

10 / 5 = 2

10 / 3 = 3.333333

10 % 3 = 1

```python
print(10/5)
print(10%5)
print(10/3)
print(10%3)
```

```python
num = int(input("Enter a # to check if it's odd or even: "))

if num % 2 == 0:
    print('Even')
else:
    print('Odd')
```

## Randomisation and Python Lists

```python
import random as r

num = r.randint(1,10) # 1 <= num <= 10

num_2 = r.random() # 0.0 <= num < 1.0

num_3 = r.uniform(1, 10) # 1.0 <= num <= 10.0

num_r = r.random() * 10 # 0.0 <= num < 10.0

print(num)
print(num_2)
print(num_3)
print(num_r)
```
